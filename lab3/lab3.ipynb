Tests from files

```python
test_nums = [1,2,3]
for n in test_nums:
    print(f"Test nr {n}:")
    filename = f"tests/example{n}.m"
    with open(filename, 'r') as f:
        data = f.read()
    parser = ParserMatrixAST()
    ast = parser.parse(Scanner().tokenize(data))
    ast.printTree(0)
    print("\n")
```
